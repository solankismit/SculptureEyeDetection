```python
import cv2
import numpy as np
```

```python
image = cv2.imread("data/dottedimg2.png")
```

```python
# gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
import matplotlib.pyplot as plt
# plt.imshow(image, cmap='')
# Show image in color
# plt.axis('off')
plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
```

```python
_, binary_image = cv2.threshold(image, 40, 250, cv2.THRESH_BINARY)
plt.imshow(binary_image,cmap='gray')
```

```python
contours, _ = cv2.findContours(binary_image, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
```

```python
min_contour_area = 10  # Minimum size of the contour to consider
filtered_contours = [contour for contour in contours if cv2.contourArea(contour) > min_contour_area]
```

```python
dot_centers = []
for contour in filtered_contours:
    M = cv2.moments(contour)
    cx = int(M["m10"] / M["m00"])
    cy = int(M["m01"] / M["m00"])
    dot_centers.append((cx, cy))
```

```python
for center in dot_centers:
    print(f"Black dot center: {center}")
```
